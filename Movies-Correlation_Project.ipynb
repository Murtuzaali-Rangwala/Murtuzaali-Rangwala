```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# reading the file
df = pd.read_csv(r"D:\Data Analyst Portfolio\Python Project\DataSet\Movies\movies1.csv")
# Exploring Data
df.head()
```

```python
 # Missing Values
for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing))
```

```python
# Data Types
df.dtypes
# We can see we have 3 types : object, int, float
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64') 
# Replace categorical values with numerical values using Label Encoding
# This creates a numerical representation for each unique genre.
df['genre'] = pd.factorize(df['genre'])[0]
df['genre'] = df['genre'].astype('int64')
```

```python
df.head()
```

```python
# After exploring, it was observed there is a difference in year and release(year)
# To resolve it, creating a new column named "year_correct"
df['year_correct'] = df['released'].astype(str).str[:4]
df.head()
```

```python
# Ordering/Sorting the data by gross revenue
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Dropping Missing Values
df.drop_duplicates()
```

```python
# Finding the correlation

# Scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'], color='#ae5956')
plt.title("Budget vs Gross")
plt.xlabel("Gross Earning")
plt.ylabel("Budget")
plt.show()
```

```python
# Plotting Budget vs Gross using seaborn
# Regression 
sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"black"})
plt.show()
```

```python
# Finding Correlation
# Attempt to convert 'budget' and 'gross' columns to numeric
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')
df = df.drop(columns=['Unnamed: 14'])
df = df.drop(columns=['Unnamed: 15'])
            
correlation_matrix = df.corr(numeric_only=True, method='pearson')
correlation_matrix
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")
plt.show()
```

```python
df_numerized_data = df
for col_name in df_numerized_data.columns:
    if(df_numerized_data[col_name].dtype == "object"):
        df_numerized_data[col_name] = df_numerized_data[col_name].astype('category')
        df_numerized_data[col_name] = df_numerized_data[col_name].cat.codes
df_numerized_data[col_name]        
    
```

```python
correlation_matrix = df_numerized_data.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")
plt.show()
```

```python
correlation_matrix = df_numerized_data.corr()
correlattion_pairs = correlation_matrix.unstack()
correlattion_pairs
```

```python
sorted_pair = correlattion_pairs.sort_values()
sorted_pair
```

```python
high_correlation = sorted_pair[(sorted_pair) > 0.5]
high_correlation
```
